```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=50
```

```python
X=np.random.rand(50)*10
X
```

```python
y=3*X+5+np.random.normal(loc=0, scale=2,size=X.shape)
y
```

```python
plt.scatter(X,y,c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
m=(N*np.sum(np.dot(X,y))-np.sum(X)*np.sum(y))/(N*np.sum(np.dot(X,X.T))-(np.sum(X)**2))
c=(np.sum(y)*np.sum(np.dot(X,X.T))-(np.sum(X)*np.sum(np.dot(X,y))))/(N*np.sum(np.dot(X,X.T))-(np.sum(X)**2))
m,c
```

```python
plt.plot(X,m*X+c)
plt.scatter(X,y,c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
