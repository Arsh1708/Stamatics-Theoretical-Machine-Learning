```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50)*10
```

```python
print(x)
```

```python
y = 3*x + 5 + np.random.normal(0,2,50)
```

```python
y
```

```python
plt.scatter(x,y,s=20,c='green')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.title('Plot of Y v/s X')
plt.show
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print(x_mean)
```

```python
slope = np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
print(slope)
```

```python
intercept = y_mean - x_mean*slope
print(intercept)
```

```python
line = slope*x + intercept
plt.plot(x,line,c='red',label='Regression Line')
plt.scatter(x,y,c='green',s=25)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot and The Regression Line')
plt.show
```
