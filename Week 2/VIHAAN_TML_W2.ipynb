```python
import numpy as np
import matplotlib.pyplot as plt
samples = 50
```

```python
x = np.random.uniform(0,10,samples)
e = np.random.normal(0,2, samples)
y = 3*x+5+e
plt.scatter(x,y)
```

```python
m = (samples*(np.dot(x,y)) - np.sum(x)*np.sum(y))/(samples*np.dot(x,x)-np.sum(x)**2)
c = (np.sum(y)*np.dot(x,x)-np.sum(x)*np.dot(x,y))/(samples*np.dot(x,x)-np.sum(x)**2)
print(m,c)
```

```python
plt.scatter(x,y,c='r')
plt.plot([np.min(x),np.max(x)], [m*np.min(x)+c,m*np.max(x)+c])
```
