```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate 50 samples for x uniformly from 0 to 10
x = np.random.uniform(0, 10, 50)
# Generate random noise epsilon from a normal distribution with mean 0 and standard deviation 2
epsilon = np.random.normal(0, 2, 50)
# Calculate y using the formula y = 3x + 5 + epsilon
y = 3 * x + 5 + epsilon
print("x samples:")
print(x)
print("\ny samples:")
print(y)
```

```python
#regression model
N = len(x)
m = (N*sum(x*y) - sum(x) * sum(y))/(N * sum(x**2) - (sum(x))**2)
c = (sum(y) * sum(x**2) - sum(x) * sum(x*y))/(N * sum(x**2) - (sum(x))**2)
print("m = ",m)
print("c = ",c)
```

```python
plt.scatter(x, y, color = "red")
X_1 = np.min(x)
Y_1 = m * X_1 + c
X_2 = np.max(x)
Y_2 = m * X_2 + c
plt.plot([X_1, X_2], [Y_1, Y_2], color = "black")
plt.legend(["Actual Data", "Regression Line"])
plt.title("y " + r'$\approx$ '  + str(round(m, 2)) + " x + " + str(round(c, 2)))
plt.show()
```
