```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 50)
epsilon = np.random.rand(50) * 2
Y = 3*X + 5 + epsilon
```

```python
plt.scatter(X, Y, color='yellow', edgecolor='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Y vs X')
plt.show()
```

```python
m = (50 * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (50 * np.sum(X ** 2) - np.sum(X) ** 2)
c = ( np.sum(Y) * np.sum(X ** 2) - np.sum(X) * np.sum(X * Y)) / (50 * np.sum(X ** 2) - np.sum(X) ** 2)
```

```python
plt.plot(X, m*X+c, color='blue', linewidth=3)
plt.scatter(X, Y, color='yellow', edgecolor='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Y vs X')
plt.show()
```
