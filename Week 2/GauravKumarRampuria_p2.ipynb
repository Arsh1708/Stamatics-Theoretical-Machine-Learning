```python
import numpy as np
import matplotlib.pyplot as plt
```

# Defining x, y, and epsilon

```python
x = np.random.rand(50)*10   # Declaring 50 samples, with range (0, 10)
epsilon = np.random.normal(0, 2, 50)    # N(0, 2) with 50 samples
y = 3*x + 5 + epsilon
```

```python
x, epsilon, y   # Displaying values of x, epsilon and y
```

# Plotting the points

```python
plt.scatter(x, y, color="yellow", edgecolor="black")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Y = 3X + 5 + e")
plt.show()
```

# Calculating m and c

```python
N = len(x)
m = (N*sum(x*y) - sum(x)*sum(y)) / (N*sum(x**2) - sum(x)**2)
c = (sum(y)*sum(x**2) - sum(x)*sum(x*y)) / (N*sum(x**2) - sum(x)**2)
```

```python
m, c
```

# Plotting the Best Fit line

```python
plt.scatter(x, y, color="yellow", edgecolor="black")
X1 = np.min(x)
Y1 = m*X1 + c
X2 = np.max(x)
Y2 = m*X2 + c
plt.plot([X1, X2], [Y1, Y2], color="blue")  # Plotting a straight line from (X1, Y1) to (X2, Y2)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['Actual Data', 'Regression Line'])
plt.title(f"Y = {round(m, 2)}X + {round(c, 2)}")
plt.show()
```
