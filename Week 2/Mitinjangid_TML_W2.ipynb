```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,10,50)
```

```python
y = 3*x + 5 +np.random.uniform(0, 2, 50)
```

```python
plt.figure(figsize=(4,4))
plt.scatter(x,y,label = 'DATA_Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('y = 3*x + 5 +N(0,2)')
plt.legend()
```

Finding best fit line:

```python
N = x.shape[0]
m = (N*sum(x*y) - sum(x) * sum(y))/(N * sum(x**2) - (sum(x))**2)
c = (sum(y) * sum(x**2) - sum(x) * sum(x*y))/(N * sum(x**2) - (sum(x))**2)
print(m)
print(c)
```

```python
reg_y = m*x + c
```
