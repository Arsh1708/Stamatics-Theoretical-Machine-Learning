```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = np.random.uniform(0, 10, 50)
print(x)
```

```python
epsilon = np.random.normal(0, 2, 50)
y = 3 * x + 5 + epsilon
print(y)
```

```python
N=50
#slope and intercept formula
m = (N*sum(x*y) - sum(x)*sum(y)) / (N*sum(x**2) - sum(x)**2)
c = (sum(y)*sum(x**2) - sum(x)*sum(x*y)) / (N*sum(x**2) - sum(x)**2)
plt.scatter(x, y, label='Data Points')
plt.plot(x, m*x + c, color='red', label='Best Fit Line')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
print(f"Slope: {m:.2f}")
print(f"Intercept: {c:.2f}")
```
