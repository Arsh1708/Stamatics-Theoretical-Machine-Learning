```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = np.random.randint(0, 10, 50)
y = 3*x + 5 +np.random.normal(0, 2, 50)
```

```python
sx = np.sum(x)
sy = np.sum(y)
sxy = np.sum(x*y)
sx2 =np.sum(x*x)

m = (50 * sxy -sx * sy)/(50 * sx2 -sx * sx)
c = (sy * sx2 - sx * sxy)/(50 * sx2 - sx * sx)
```

```python
plt.scatter(x, y)
plt.plot(x, m*x + c, color = "red")
plt.show()
print("LINE IS: y = ", m, "x + ", c)
```
