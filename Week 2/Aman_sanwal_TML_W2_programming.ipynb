```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(50)*10
```

```python
X
```

```python
Y = 3*X + 5 + np.random.normal(0, 2, 50)
```

```python
plt.scatter(X, Y, c='blue', s=30)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Scatter plot of y vs x')
plt.show
```

```python
N=50
m = (N*sum(X*Y) - sum(X) * sum(Y))/(N * sum(X**2) - (sum(X))**2)
c = (sum(Y) * sum(X**2) - sum(X) * sum(X*Y))/(N * sum(X**2) - (sum(X))**2)
print(f"The value of m is: {m}\nThe value of c is {c}")
```

```python
plt.scatter(X, Y, color = "red", s=30)
X_1 = np.min(X)
Y_1 = m * X_1 + c
X_2 = np.max(X)
Y_2 = m * X_2 + c
plt.plot([X_1, X_2], [Y_1, Y_2], color = "black")
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Scatter plot of y vs x')
plt.show()
```
