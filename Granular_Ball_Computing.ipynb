```python
def kmeans(X_train,y_train,balls):
    # Randomly initialize two distinct centroids
    indices = random.sample(range(len(X_train)), 2) 
    centroid1 = X_train.iloc[indices[0]].values 
    centroid2 = X_train.iloc[indices[1]].values  

    # Helper function to calculate Euclidean distance
    def calculate_distance(point, centroid):
        return np.linalg.norm(point - centroid)

    # Run 2-means algorithm
    max_iterations = 100
    for iteration in range(max_iterations):
        cluster1 = []  # Stores indices of points assigned to centroid1
        cluster2 = []  

        for i, point in X_train.iterrows(): 
            point = point.values  
            distance_to_centroid1 = calculate_distance(point, centroid1)
            distance_to_centroid2 = calculate_distance(point, centroid2)
            
            if distance_to_centroid1 < distance_to_centroid2:
                cluster1.append(i) 
            else:
                cluster2.append(i)  

        # Update centroids
        new_centroid1 = X_train.iloc[cluster1].mean(axis=0).values if cluster1 else centroid1
        new_centroid2 = X_train.iloc[cluster2].mean(axis=0).values if cluster2 else centroid2

        centroid1, centroid2 = new_centroid1, new_centroid2

    # Collect indices and corresponding labels
    cluster1_with_labels = [(i, y_train.iloc[i]) for i in cluster1]
    cluster2_with_labels = [(i, y_train.iloc[i]) for i in cluster2]

    # print("Cluster 1 (indices and labels):", cluster1_with_labels[:6])
    # print("Cluster 1 (indices and labels):", cluster2_with_labels[:6])
    # print(len(cluster1_with_labels))
    # print(len(cluster2_with_labels))

    #_____________________________________________

    indices1 = [i for i, label in cluster1_with_labels]
    center1 = np.zeros(X_train.shape[1])
    for i in indices1:
        center1 += X_train.iloc[i].values
    center1 /= len(indices1)
    # print(f"Center of Gravity (Centroid) of Cluster 1: {center1}")
    distances = [np.linalg.norm(X_train.iloc[i].values - center1) for i in indices1]
    average_radius = np.mean(distances)
    print(type(average_radius))
    # print(f"Average Radius of Cluster 1: {average_radius}")
    labels=[int(y_train.iloc[i].values) for i in indices1]
    count1=labels.count(0)
    count2=labels.count(1)
    purity = max(count1, count2) / len(labels)
    # print(f"Purity of Cluster 1: {purity}")


    indices2 = [i for i, label in cluster2_with_labels]
    center2 = np.zeros(X_train.shape[1])
    for i in indices2:
        center2 += X_train.iloc[i].values
    center2 /= len(indices2)
    # print(f"Center of Gravity (Centroid) of Cluster 2: {center2}")
    distances = [np.linalg.norm(X_train.iloc[i].values - center2) for i in indices2]
    average_radius2 = np.mean(distances)
    # print(type(average_radius))
    # print(f"Average Radius of Cluster 2: {average_radius2}")
    labels2=[int(y_train.iloc[i].values) for i in indices2]
    count3=labels2.count(0)
    count4=labels2.count(1)
    # print(labels)
    purity2 = max(count3, count4) / len(labels2)
    # print(f"Purity of Cluster 2: {purity2}")
    # print(type(average_radius))
    balls.append([purity,float(average_radius),center1,indices1,labels])
    balls.append([purity2,float(average_radius2),center2,indices2,labels2])

```

```python
def kmeans_recursive(X_train, y_train, balls, purity_threshold):
    # Randomly initialize two distinct centroids
    indices = random.sample(range(len(X_train)), 2)
    centroid1 = X_train.iloc[indices[0]].values
    centroid2 = X_train.iloc[indices[1]].values

    # Helper function to calculate Euclidean distance
    def calculate_distance(point, centroid):
        return np.linalg.norm(point - centroid)

    # Run 2-means algorithm
    max_iterations = 100
    for iteration in range(max_iterations):
        cluster1 = []  # Stores indices of points assigned to centroid1
        cluster2 = []

        for i, point in X_train.iterrows():
            point = point.values
            distance_to_centroid1 = calculate_distance(point, centroid1)
            distance_to_centroid2 = calculate_distance(point, centroid2)

            if distance_to_centroid1 < distance_to_centroid2:
                cluster1.append(i)
            else:
                cluster2.append(i)

        # Update centroids
        new_centroid1 = X_train.iloc[cluster1].mean(axis=0).values if cluster1 else centroid1
        new_centroid2 = X_train.iloc[cluster2].mean(axis=0).values if cluster2 else centroid2

        centroid1, centroid2 = new_centroid1, new_centroid2

    # Collect indices and corresponding labels
    cluster1_with_labels = [(i, y_train.iloc[i]) for i in cluster1]
    cluster2_with_labels = [(i, y_train.iloc[i]) for i in cluster2]

    # Calculate purity, center, and radius for Cluster 1
    indices1 = [i for i, label in cluster1_with_labels]
    center1 = np.mean(X_train.iloc[indices1].values, axis=0) if indices1 else np.zeros(X_train.shape[1])
    distances1 = [np.linalg.norm(X_train.iloc[i].values - center1) for i in indices1]
    average_radius1 = np.mean(distances1) if distances1 else 0
    labels1 = [int(y_train.iloc[i]) for i in indices1]
    count1_0 = labels1.count(0)
    count1_1 = labels1.count(1)
    purity1 = max(count1_0, count1_1) / len(labels1) if labels1 else 1

    # Calculate purity, center, and radius for Cluster 2
    indices2 = [i for i, label in cluster2_with_labels]
    center2 = np.mean(X_train.iloc[indices2].values, axis=0) if indices2 else np.zeros(X_train.shape[1])
    distances2 = [np.linalg.norm(X_train.iloc[i].values - center2) for i in indices2]
    average_radius2 = np.mean(distances2) if distances2 else 0
    labels2 = [int(y_train.iloc[i]) for i in indices2]
    count2_0 = labels2.count(0)
    count2_1 = labels2.count(1)
    purity2 = max(count2_0, count2_1) / len(labels2) if labels2 else 1

    # Append granular balls to the list
    ball1 = None
    ball2 = None
    if indices1:
        ball1 = [purity1, float(average_radius1), center1, indices1,labels1]
        balls.append(ball1)
    if indices2:
        ball2 = [purity2, float(average_radius2), center2, indices2,labels2]
        balls.append(ball2)

    if len(indices1)>80 and purity1 < purity_threshold  and indices1:
        X_sub = X_train.iloc[indices1].reset_index(drop=True) 
        y_sub = y_train.iloc[indices1].reset_index(drop=True) 
        balls.remove(ball1)  # Remove the ball after entering the recursive call
        kmeans_recursive(X_sub, y_sub, balls, purity_threshold)

    if len(indices2)>80 and purity2 < purity_threshold  and indices2:
        X_sub = X_train.iloc[indices2].reset_index(drop=True)  
        y_sub = y_train.iloc[indices2].reset_index(drop=True)  
        balls.remove(ball2)  # Remove the ball after entering the recursive call
        kmeans_recursive(X_sub, y_sub, balls, purity_threshold)

```
